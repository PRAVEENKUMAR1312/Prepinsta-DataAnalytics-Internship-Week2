```python
def display_game(positions):
    print("[ 0 ,  1 ,  2 ,  3 ,  4 ,  5 ,  6 ]")
    print(positions)


def frog_leap_game():
    positions = ['G', 'G', 'G', '-', 'B', 'B', 'B']

    while True:
        display_game(positions)

        # Take input from the user
        pos = input("Press q to quit\nEnter position of piece: ")

        if pos == 'q':
            print("You Lose")
            break
        else:
            pos = int(pos)

        # Check validity of the selected position
        if pos < 0 or pos > 6 or positions[pos] == '-':
            print("Invalid Move")
            continue

        pos2 = 0  # Initialize pos2

        if positions[pos] == 'G':
            # For 'G' frog
            if pos + 1 <= 6 and positions[pos + 1] == '-':
                pos2 = pos + 1
            elif pos + 2 <= 6 and positions[pos + 2] == '-' and positions[pos + 1] == 'B':
                pos2 = pos + 2
            else:
                print("Invalid Move")
        elif positions[pos] == 'B':
            # For 'B' frog
            if pos - 1 >= 0 and positions[pos - 1] == '-':
                pos2 = pos - 1
            elif pos - 2 >= 0 and positions[pos - 2] == '-' and positions[pos - 1] == 'G':
                pos2 = pos - 2
            else:
                print("Invalid Move")

        # Make the move by swapping elements
        positions[pos], positions[pos2] = positions[pos2], positions[pos]

        # Check for winning condition
        if positions == ['B', 'B', 'B', '-', 'G', 'G', 'G']:
            display_game(positions)
            print("You Win")
            break


# Run the Frog Leap game
frog_leap_game()
```
